```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Hàm đọc một ảnh
def read_image(filename):
    img = Image.open(filename)
    img = img.convert('L')
    img = np.array(img)
    return img
```

```python
file_name = "Image/Fig0309(a)(washed_out_aerial_image).tif"
old = read_image(file_name) 
plt.imshow(old, cmap="gray")
```

```python
def slice_image(image, x_coor, y_coor, x_size, y_size):
    return image[y_coor:y_coor+y_size, x_coor:x_coor+x_size]
```

```python
y_coor = 500
x_coor = 400
y_size = 150
x_size = 130
cropped= slice_image(old, x_coor, y_coor, x_size, y_size)
plt.imshow(cropped, cmap="gray")
```

```python
# slice image with height = 150 and width = 130
def crop_size_image(image, width_new, height_new):
    return image[:height_new, :width_new]
```

```python
cropped= crop_size_image(old, 100, 200)
plt.imshow(cropped, cmap="gray")
```
