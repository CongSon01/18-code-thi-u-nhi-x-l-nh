```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

# perform a fast fourier transform and create a scaled, frequency transform image
def ft_image(norm_image):
   '''This function takes in a normalized, grayscale image
       and returns a frequency spectrum transform of that image. '''
   ft = np.fft.ifftshift(norm_image)
   ft = np.fft.fft2(ft)
   ft = np.fft.fftshift(ft)
   frequency_tx = np.log(abs(ft))
    
   return frequency_tx

def ft_inverse(norm_image):
    ift = np.fft.ifftshift(norm_image)
    ift = np.fft.ifft2(ift)
    ift = np.fft.fftshift(ift)
    ift = ift.real  # Take only the real part
    return ift
```

```python

import numpy as np
import matplotlib.pyplot as plt
# Read in the images
# image  = cv2.imread('Image/Fig0354(a)(einstein_orig).tif')
image  = cv2.imread('Image/stripes.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

# normalize the image
norm_image = gray/255.0

f_image = ft_image(norm_image)

f_inverse = ft_inverse(norm_image)

# Display the images
f, (ax1,ax2, ax3, ax4, ax5) = plt.subplots(1,5, figsize=(20,10))

ax1.set_title('original image')
ax1.imshow(image)

ax2.set_title('gray')
ax2.imshow(gray, cmap='gray')

ax3.set_title('normalized image')
ax3.imshow(norm_image)

ax4.set_title('frequency Fourier  transform ')
ax4.imshow(f_image, cmap='gray')

ax5.set_title('Inverse Fourier transform ')
ax5.imshow(f_inverse, cmap='gray')
```

```python
def read_image(filename):
    image = Image.open(filename)
    image = image.convert('L')
    return image
```

```python
def fourier_transfer(image):
    ft = np.fft.ifftshift(image)
    ft = np.fft.fft2(ft)
    fourier_image = np.fft.fftshift(ft)
    return fourier_image
```
