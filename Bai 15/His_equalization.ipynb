```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image 
from skimage import io
import matplotlib.image as im
img = im.imread('../Image/Fig0320(2)(2nd_from_top).tif')

# pr so diem anh
pr, bins = np.histogram(img, bins = 256, range=(0,256))
# mat do
cr = pr / (img.shape[0] * img.shape[1])
# bins[:-1] bo khoang cuoi
plt.plot(bins[:-1], pr)
```

```python
plt.imshow(img, cmap='gray')
```

```python
# Tinh phan phoi tich luy
sum = 0
for i in range(cr.shape[0]):
    sum += cr[i]
    cr[i] = sum
plt.plot(bins[:-1], cr)
```

```python
s = np.empty_like(256)
s = (256 - 1) * cr
s = s.astype('uint8')
```

```python
s
```

```python
equalized_img = np.empty_like(img)

for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        equalized_img[i][j] = s[img[i][j]]

plt.imshow(equalized_img, cmap='gray')
```

```python
# new histogram
pr_new, bins_new = np.histogram(equalized_img, bins = 256, range=(0,256))
plt.plot(bins_new[:-1], pr_new)
```
