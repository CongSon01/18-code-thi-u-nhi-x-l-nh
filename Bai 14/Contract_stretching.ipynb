```python
# -*- coding: utf-8 -*-
"""
giãn ảnh :
    Sự giãn nở hình thái làm cho các vật thể dễ nhìn thấy hơn và lấp đầy các
    lỗ nhỏ trên vật thể. Các đường kẻ có vẻ dày hơn và các hình được tô có
    vẻ lớn hơn.
"""

import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as img

image = img.imread('../Image/Fig0354(a)(einstein_orig).tif')

plt.hist(image)
plt.show()

def adjust_image_gamma(image, gamma = 0.1):
  image = np.power(image, gamma)
  max_val = np.max(image.ravel())
  image = image/max_val * 255
  image = image.astype(np.uint8)
  return image

low_adjusted = adjust_image_gamma(image, 0.1)
plt.imshow(low_adjusted)
```

```python
histogram, bin_edges = np.histogram(image, bins=256, range=(0, 255))
plt.figure()
plt.title("Grayscale Histogram")
plt.xlabel("grayscale value")
plt.ylabel("pixel count")
plt.plot(histogram) 
plt.show()
```

```python
# create the histogram
histogram_1, bin_edges_1 = np.histogram(low_adjusted, bins=256, range=(0, 255))
# configure and draw the histogram figure
plt.figure()
plt.title("Grayscale Histogram")
plt.xlabel("grayscale value")
plt.ylabel("pixel count")

plt.plot(bin_edges_1[0:-1], histogram_1)  # <- or here
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image 
image = Image.open("../Image/Fig0354(a)(einstein_orig).tif")
fig, ax = plt.subplots()
plt.imshow(image, cmap='gray')
plt.show()
```

```python
histogram, bin_edges = np.histogram(image, bins=256, range=(0, 255))
plt.figure()
plt.title("Grayscale Histogram")
plt.xlabel("grayscale value")
plt.ylabel("pixel count")
plt.plot(histogram) 
plt.show()
```

```python
def contract_stretching(image):
    a = 0
    b = 255
    c = np.max(image)
    d = np.min(image)
    
    # im_out = np.zeros(image.shape)
    im_out = (image - c) * ((a - b) / (c - d)) + a
    return im_out
```

```python
# create image ouput with contrast stretching
# image_out = skimage.exposure.rescale_intensity(image, out_range=(0, 1))
image_out = contract_stretching(image)
fig, ax = plt.subplots()
plt.imshow(image_out, cmap='gray')
plt.show()
```

```python
# create the histogram
histogram_1, bin_edges_1 = np.histogram(image_out, bins=256, range=(0, 255))
# configure and draw the histogram figure
plt.figure()
plt.title("Grayscale Histogram")
plt.xlabel("grayscale value")
plt.ylabel("pixel count")

plt.plot(bin_edges_1[0:-1], histogram_1)  # <- or here
plt.show()
```
